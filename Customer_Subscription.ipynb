```python
plt.figure(figsize=(16,9))
features=data.columns
for i,j in enumerate(features):
  plt.subplot(3,3,i+1)
  plt.title("Histogram of {}".format(j),fontsize=15)

  bins = len(data[j].unique())
  plt.hist(data[j],bins = bins,rwidth=0.8,edgecolor="y",linewidth=2)
plt.subplots_adjust(hspace=0.5)
```

```python
age = data['age']
bins=[10,20,30,40,50,60,70]
plt.hist(age,bins,edgecolor='black',log=True)
plt.axvline(x=35,color='blue')
```

```python
age_x=data['age']
hour_y=data['hour']
plt.scatter(age_x,hour_y,s=40,alpha=0.9)
```

```python
plt.style.use('ggplot')
plt.plot(age_x,hour_y,'y',linestyle='--',marker='o',label='india')

plt.grid()
plt.legend()
```

```python

plt.plot(age_x,hour_y)        #plotting age_x on 'x' axis & height_y on 'y' axis
plt.title('hour vs age')      #giving the title to this whole curve
plt.xlabel('age')               #giving a name to 'x' axis
plt.ylabel('hour')
```

```python
dayofweek=data['dayofweek']

plt.plot(age_x,hour_y)        #plotting age_x on 'x' axis & height_y on 'y' axis
plt.plot(age_x,dayofweek)   #giving the title to this whole curve
plt.legend(['hour','dow'])
```

```python
plt.style.use('ggplot')
plt.plot(age_x,hour_y,'y',linestyle='--',marker='o',label='india')
plt.bar(age_x,dayofweek,color='k',label='us')                             #plotting bar charts
plt.grid()
plt.legend()
```

```python
plt.style.use('ggplot')
plt.bar(age_x,hour_y,color='#444444')
plt.bar(age_x,dayofweek,color='#008fd5')             #here two bar charts are coincided
plt.grid(True)
plt.legend()
```
